```python
import pyTSPA
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pyTSPA.load_match_data("Football_ENG_2017_18.xlsx")
df
```

```python
# Generate necessary metrics
team_performance_df = pyTSPA.each_team_performance(df)
win_percentage_df = pyTSPA.each_win_percentage(df)
pythagorean_expectation_df = pyTSPA.each_pythagorean_expectation(df)
result_stats_dict = pyTSPA.result_stats(df)

# Plot Result Distribution
print("Plotting result distribution")
pyTSPA.plot_result_distribution(df)

# Plot Team Results for a specific team
print("Plotting team results for 'Team A'")
pyTSPA.plot_team_results(df, 'Team A')

# Plot League Points Table
print("Plotting league points table")
pyTSPA.plot_league_points_table(team_performance_df)

# Plot Goal Difference Distribution
print("Plotting goal difference distribution")
pyTSPA.plot_goal_difference_distribution(team_performance_df)

# Plot Win Percentage Comparison
print("Plotting win percentage comparison")
pyTSPA.plot_win_percentage_comparison(win_percentage_df)

# Plot Pythagorean Expectation
print("Plotting pythagorean expectation")
team_performance_df = team_performance_df.merge(pythagorean_expectation_df, on='Team', how='left')
pyTSPA.plot_pythagorean_expectation(team_performance_df)
```
